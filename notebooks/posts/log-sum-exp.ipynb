```python
# Trivial example to test the binary cross entropy loss

y_true = torch.tensor([0., 1., 1., 0.])
logits = torch.tensor([0.1, 0.9, 0.8, 0.2])

loss = torch.nn.BCEWithLogitsLoss()
loss(logits, y_true)
```

```python

y_hat = torch.nn.Sigmoid()(logits)
y_hat
```

```python
# BCE
-(y_true * torch.log(y_hat) + (1 - y_true) * torch.log(1 - y_hat)).mean()
```

```python
def our_bce(y_true, logits):
    y_hat = torch.nn.Sigmoid()(logits)
    return -(y_true * torch.log(y_hat) + (1 - y_true) * torch.log(1 - y_hat)).mean()
```

```python
# check that our implementation is correct
assert our_bce(y_true, logits) == loss(logits, y_true)
```

```python
# Larger logits
logits = torch.tensor([100., 200., 300., 400.])

assert our_bce(y_true, logits) == loss(logits, y_true)

```
